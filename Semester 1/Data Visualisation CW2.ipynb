## Research question 1: Does parents’ education levels affect children’s likelihood to attend university?


According to a German study[1], a young person's family background and the education of his or her parents have a significant impact on the educational chances available to him or her. 
The study compares children whose parents did not both graduate from university (group A) to children whose parents did graduate from university (group B) (group B).

If 100 students from each group enter elementary school, only 21 from group A will attend university, while 74 from group B will begin their studies.However, the difference does not end there. Only 15 of Group A's 21 university students earn a bachelor's degree (equivalent to an undergraduate course/degree in the United Kingdom), and only 8 of those 15 earn a master's degree (equals postgraduate in the UK).

In contrast, 63 of the 74 children in Group B who began university have completed their bachelor's degree, with 45 completing a master's degree.Only one of the 100 children in group A receives a Ph.D., whereas ten children in group B receive a Ph.D. 

A research by the National Center for Education Statistics[2] (NCES) in the United States found that children whose parents went to college are substantially more likely to go to college and graduate. Clearly, the research presented suggest that a parent's degree of education influences their children's likelihood of attending university, but does this hold true in other countries?

Unfortunately, there are no studies from other countries, so this survey was a fantastic opportunity to learn more about this topic in other nations and solidify the hypothesis for countries like Germany and the United States. I distributed this survey to my colleagues from the University of Londons's MSc Data Science program, in the hopes of receiving responses from students all over the world. ## Research question 1: Does parents’ education levels affect children’s likelihood to attend university?



```python
# setting figsize
plt.rcParams["figure.figsize"] = (15, 10)

# using crosstab to show correlation between parents_education and highest_education
table = pd.crosstab(survey_df['parents_education'],survey_df['highest_education'])

# rearanging the table
table = table[['None of the above',"General Certificate of Secondary Education(High School Degree)","Bachelor's degree(Undergraduate degree)","Master's degree(Postgraduate degree)","PhD"]]
```

```python
# plotting bar charts side by side for comarison
index=['No Education','High School','Bachelors','Masters','PhD']
plotdata = pd.DataFrame({'Education':list(table.iloc[1]) ,
                   'No Education': list(table.iloc[0])},index=index)

ax = plotdata.plot.bar()
plt.legend(fontsize=12)
ax.set_title('Highest education of respondents by education of parents')
ax.set_xlabel('Education')
ax.set_ylabel('Amount of people')
```

At first glance it seems like the hypothesis has been confirmed. However, there were way more students among survey respondents to make a confident call about wether the hypothesis can be confirmed. Some further investigation is needed.

```python
# getting statistics of last table for pie charts
all1 = sum(table.iloc[1])
all2 = sum(table.iloc[0])
University_percentage1 =(table.iat[1,2]+table.iat[1,3]+table.iat[1,4])
University_percentage2 =(table.iat[0,2]+table.iat[0,3]+table.iat[0,4]) 
rest1=all1-University_percentage1
rest2=all2-University_percentage2
```

```python
# plotting pie charts side by side with percentage of  each  group
labels = ['at university', 'not at university']
sizes1 = [University_percentage1, rest1]
sizes2=[University_percentage2, rest2]

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.pie(sizes1,   autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax2.pie(sizes2,   autopct='%1.1f%%', startangle=90)
ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(fontsize=12,labels=labels)
ax1.set_title('% of people who made  it to university if the one of the parents did')
ax2.set_title('% of people who made it to university if none of the parents did')
```

So I plotted those pie charts. On the left is the pie chart for all the people with at least one parent having higher education. A remarkable 92% of them made it to university themselver. On the right is the pie chart for people who do not have highly educated parents. Only 66% of them  made it college. The  sample  size  is low but this  comparison shows that children  from parents with academic backround are more likely to attend university themselves. Also the percentage on the right might be inflated, because nowadays more and more people  go  to university.

## Research question 2: What impact did distance learning, caused by the pandemic, have on student performance?



```python
survey_df.online_study.value_counts()
```

I will investigate 2 different groups: Group 1, which transitioned into distance learning and is currently online learning and group 2, which is neither of those things. I seperated them because for some people it might be the preffered way to study/attend university so that has to be taken into account.

```python
# get group1 that transitioned into online learning and is/was still learning online
# get group 2 that did not transition into online learning and is still offline learning
group1=survey_df.loc[(survey_df['online_study']=='Yes')  & (survey_df['online_study_transition'] == 'Yes')]
group2=survey_df.loc[(survey_df['online_study']=='No')  & (survey_df['online_study_transition'] == 'No')]
```

```python
group1.reset_index(drop=True)
```

```python
# crosstabing by gender and  online study impact, with margins all
table1 = pd.crosstab(group1['gender'],group1['online_study_impact'],margins=True)
# new order for columns
table1 = table1[['Grades got worse','Grades stayed the same','Grades got better']]
table1  = table1.iloc[2]
```

```python
# plotting bar chart for group 1
# setting figsize
plt.rcParams["figure.figsize"] = (12, 7)
ax = table1.plot.bar()
ax.set_title('grade progression according to students from group 1')
ax.set_xlabel('grades from getting worse to getting better')
ax.set_ylabel('number of reported sentiments')
plt.show()
```

```python
# crosstabing by gender and  online study impact, with margins all
table2 = pd.crosstab(group2['gender'],group2['online_study_impact'],margins=True)
# new order for columns
table2 = table2[['Grades got worse','Grades stayed the same','Grades got better']]
#slicing of irrelevant rows
table2  = table2.iloc[3]
table2
```

```python
# plotting bar chart for group 2
ax = table2.plot.bar()
ax.set_title('grade progression according to students from group 2')
ax.set_xlabel('grades from getting worse to getting better')
ax.set_ylabel('number of reported sentiments')
plt.show()
```

```python
# plotting bar charts side by side for comparison for each group
index=['Grades got worse','Grades stayed the same','Grades got better']

plotdata = pd.DataFrame({'group1':list(table1) ,
                   'group2': list(table2)},index=index)

ax = plotdata.plot.bar()
plt.legend(fontsize=12)
ax.set_title('comparison  of grade progression by groups')
ax.set_xlabel('grades from getting worse to getting better')
ax.set_ylabel('number of reported sentiments')
```

I intentionally split the data for this research question into 2 groups:
    1. Group1: Students  who  study  online right now and transitioned into online  learning  during the pandemic
    2. Group2: Students  who do not study online right now and have not transitioned
    
I  split those groups because some people might have the preference to study online and like it more anyway.

Most people said that grades stayed the same with a good chunk of respondents saying that grades got worse. For group2, the not online group, grades rarely got better. Maybe the grades for group1 got worse because they had less social interaction as they already had, isolating completly as they do not have any social gatherings at university since they study online.

```python
# crosstabing by gender and  online study impact, with margins all
table3 = pd.crosstab(group1['gender'],group1['online_study_satisfaction'],margins=True)
# new order for columns
table3 = table3[['Very dissatisfied','Dissatisfied','Neutral','Satisfied','Very Satisfied']]
table3  = table3.iloc[2]

#repeating the process for group2
# crosstabing by gender and  online study impact, with margins all
table4 = pd.crosstab(group2['gender'],group2['online_study_satisfaction'],margins=True)
# new order for columns
table4 = table4[['Very dissatisfied','Dissatisfied','Neutral','Satisfied','Very Satisfied']]
table4  = table4.iloc[3]
```

```python
# plotting bar charts side by side for comparison
index=['Very dissatisfied','Dissatisfied','Neutral','Satisfied','Very Satisfied']

plotdata = pd.DataFrame({'group1':list(table3) ,
                   'group2': list(table4)},index=index)

ax = plotdata.plot.bar()
plt.legend(fontsize=12)
ax.set_title('comparison  of satistfaction with e-learning by groups')
ax.set_xlabel('sentiment from bad to good')
ax.set_ylabel('number of reported sentiments')
```

# Conclusions

### Research question 1: Does parents’ education levels affect children’s likelihood to attend university?

The answer is yes! It does. Even though there was bias in the survey data, because almost only students took the survey, it was clearly shown that the percentage of the group with high education parents, was way higher than for the other group.
Since more and more people go to university each year, this might have massive implications for the future. Traditional jobs like wood workers are already having problems with recruitment and if every person only wants to go to university anymore, competiton among university graduates will be fierce.

Of course this sounds way to dramatic for the small sample size that was  investigated, but the trend is still there.

### Research question 2: What impact did distance learning, caused by the pandemic, have on student performance?

Well for most students, the grades stayed the same which is good and actually unexpected. One might think that a global pandemic could influence the students in a bay way, because of stress or fear etc. For some the grades got worse, which might be because of the reasons just stated. However, there was the trend towards satisfaction among students about the e-learning opportunities provided by universities. This research question has a bigger scope than just grades and satisfaction and  intangibles like social connections during university work have not  been investigated. There were for sure some students, who were negativly impacted by distance learning as they need close contact to their peers. But the survey questions resulted in somewhat clear answers when plotted.

### Research question 3: Does a high tertiary graduation rate correlate to a high vaccination rate?

This research questions was the hardest to answer. In the research leading up to this project, some  sources made it seem  like it was  obvious that well  educated people were more likly to get vaccinated. However, I could not 100% confirm this during my analysis. And the fact that I  could not makes sense. We are still dealing with a somewhat deadly virus, so  it makes sense for people to get vaccinated, of course  regardless of their education and knowledge. On the other hand, the external data did not state wether the vaccination percentages came to be because a lot of highly educated people got vaccinated. So basically my external datasets were useless to some extent. To answer this questions now: No, a high tertiary graduation rate does not necessarily correlate to a high vaccination rate. 

### Research question 4: Are students taking extra steps to increase their employability?

The simplest questions that had to be answered in this investigation. And the answer is yes. Of course students are taking extra steps to increase their employability. And why would they not? Making an  account on LinkedIn literally takes seconds. Every job offer asks for years of experience, so it makes sense for students to try and gain working experience while they are still in university. The bar charts perfectly described this same sentiment, as the number of students taking extra steps was at least  for times  higher than the  number of students who do not.
